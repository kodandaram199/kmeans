### What is Image Segmentation and why it is done and where it is used
- The process of segmenting image into easily analysable parts is called image segmentation. 
- Most of the times all the parts of images may not contain useful information, so instead of processing entire image, we segment the image and then use it for processing. 
- It is mainly used in object detection and edge detection.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
original = cv2.imread('fruits.jpeg')
```

```python
plt.imshow(original)
```

#### opencv reads img in BGR so convert to RGB to visualize original img

```python
img=cv2.cvtColor(original,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

#### sklearn models accept only 2d input

```python
vectorized = img.reshape((-1,3))
print(vectorized.shape)
print(vectorized[0])
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(vectorized)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
center = np.uint8(kmeans.cluster_centers_)
res = center[kmeans.labels_.flatten()]
result_image = res.reshape(img.shape)
print(center[1])
print(kmeans.labels_[1])
print(center[kmeans.labels_[1]])
print(kmeans.labels_[1].flatten())
print(center[[3]])
print(center[3].shape)
print(center[[3]].shape)
print(res.shape)
```

```python
plt.imshow(result_image)
```
